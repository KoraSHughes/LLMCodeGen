# Main File To Analyise Data Yield

```python
import os
import pandas as pd
import numpy as np
```

## Testbenches

### Open Titan

```python
df = pd.read_csv("example-code/verilog examples/open-titan/asserted-files.csv").iloc[:, 1:]
df["Assertion-Count"] = df["Data"].apply(lambda code : len([line for line in code.split('\n') if 'assert' in line]))
df = df[df["Assertion-Count"] > 0]
df
```

```python
df.describe()
```

```python
total_assertions = np.sum(df["Assertion-Count"])
print("Found a total of", total_assertions, "assertions in this dataset")
```
